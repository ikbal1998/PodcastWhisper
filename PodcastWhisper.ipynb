### Collecting Podcast URLs

```python
import json
import requests
import pandas as pd
from datetime import datetime, timedelta
import datetime as dt
import time
import os


date_boundary_1 = "2022-07-20"
date_boundary_2 = "2022-08-20"


podcast_keywords = ['midas%20podcast%20aksam%20bulteni','midas%20podcast%20piyasa%20turu']

start_date = datetime.strptime(date_boundary_1, "%Y-%m-%d")
end_date = datetime.strptime(date_boundary_2, "%Y-%m-%d") + timedelta(days=1)


interval_size = 90
intervals = []
current_date = start_date
while current_date <= end_date:
    next_date = current_date + timedelta(days=interval_size - 1)
    if next_date > end_date:
        next_date = end_date
    intervals.append((int(current_date.timestamp()*1000),int(next_date.timestamp()*1000)))
    current_date = next_date + timedelta(days=1)


df_list = []
for start,end in intervals:
    for off in [0, 10]:
      for key in podcast_keywords:
        url = ('https://listen-api.listennotes.com/api/v2/search?'
              f'q={key}&'
              'sort_by_date=0&'
              'type=episode&'
              f'offset={off}&'
              'len_min=0&'
              'len_max=30&'
              'genre_ids=99%2C95&'
              f'published_before={end}&'
              f'published_after={start}&'
              #'only_in=title%2Cdescription&'
              'language=Turkish&'
              'safe_mode=0&'
              'unique_podcasts=0&'
              'page_size=10'
        )
        headers = {
          'X-ListenAPI-Key': '??????????',
        }
        time.sleep(1)
        response = requests.request('GET', url, headers=headers)
        data = response.json()
        df = pd.DataFrame(data['results'])
        df_list.append(df)


podcast_urls = pd.concat(df_list, ignore_index= True) #if len(df_list) > 0 else pd.DataFrame(columns=cols)


```

```python

audio_urls = podcast_urls[['audio','pub_date_ms']].rename(columns={'pub_date_ms':'date','audio':'url'})
date = audio_urls['date'] / 1000
formatted_date = []
for i in date:  
    ordinary_date = dt.datetime.fromtimestamp(i)
    formatted_date.append(ordinary_date.strftime("%Y-%m-%d"))
    
audio_urls['date'] = pd.Series(formatted_date)


def insert_to_csv_with_header(dataframe, file_path):
    file_exists = os.path.isfile(file_path)
    if not file_exists:
        dataframe.to_csv('podcasts_pool.csv', index=False)
    else:
        dataframe.to_csv('podcasts_pool.csv', mode='a', header=False, index=False)

insert_to_csv_with_header(audio_urls, 'podcasts_pool.csv')
```

### Transcription Model (Whisper)

```python
import whisper

transcript = {}
for row in audio_urls.itertuples(index=False):
    model = whisper.load_model("base")
    result = model.transcribe(row.url)
    transcript[row.date] = result["text"]
    
df = pd.DataFrame(list(transcript.items()), columns=['date', 'transcript'])
df['date'] = pd.to_datetime(df['date'])
df.set_index("date",inplace=True)

csv_file_path = "whisper_results.csv"
df.to_csv(csv_file_path, index=True)
```
